```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(color_codes=True) #adds a coloured backgroud and adds grid lines
%matplotlib inline
```

```python
auto = pd.read_csv("Automobile (1).csv")
```

```python
auto.head()
```

                    *PLOTS FOR CONTINUOUS VARIABLES*

```python
sns.distplot(auto['highway_mpg']);
```

*The continuous Line on the histogram integrates to be 1 and is known as the KDE - Kernel Density Estimation. The width of each bar times the height of each bar all sum up to 1*

```python
sns.distplot(auto["city_mpg"]) #gives a histogram, 
                               #distplot - seaborn's word for 
                               #histogram plot the distribution series
```

*A Rug plot turns the KDE off and also puts little ticks wherever there is a value or a data point, on x-axis*

```python
sns.distplot(auto["city_mpg"], kde=False, rug=True)
```

*Scatter Plot, visualizing a relationship between two variables using Seaborn'sjointplot() function*

```python
sns.jointplot(auto["engine_size"], auto["horsepower"]);
```

*Hex bin plots - 2D histogram, where the height is based on the colour instead of height of a bar. Darker the hexagon more number of data points are present in it*

```python
sns.jointplot(auto["engine_size"], auto["horsepower"], kind="hex");
```

*KDE JOINTPLOT : All the areas are made of joint distribution, The dark area carries the maximum of joint distribution, (KDE estimation/approximation of joint distribution)*

```python
sns.jointplot(auto["engine_size"], auto["horsepower"], kind = "kde");
```

*Pair Plot : Takes dataframe as an argument it is used to plot multiple scatterplots in a dataset, for every pair of coloum in the data frame it creates a scatter plot and for each individual coloum it makes a histogram*

```python
sns.pairplot(auto[["normalized_losses","engine_size","horsepower"]]);
```

            *PLOTS FOR CONTINUOUS AND CATEGORICAL VARIABLES*

*STRIP PLOT: Same as scatter plot with one of the axis as categories. Crowded aears on the map with blue points have more datapoints. With jitter=True we get a clarity on the data points or else it depicts a single line of dots*

```python
sns.stripplot(auto['fuel_type'], auto['horsepower'], jitter=False);
```

```python
sns.stripplot(auto['fuel_type'], auto['horsepower'], jitter=True);
```

*SWARM PLOT : Positions each scatterplot point on the categorical axis and avoids overlapping points, shows the spread of data, ie; how many datapoints r there in the area*

```python
sns.swarmplot(auto["fuel_type"], auto["horsepower"]);
```

*BOX PLOTS: Common Statistical Plot, similar to a histogram but instead of having bars for every bin there just a few specific points that are noted *

```python
sns.boxplot(auto["number_of_doors"], auto["horsepower"]);
#blue box represents the car that has 2 doors(median-100 horsepower), 
#the orange one represents the box that has 4 doors(median-90 horsepower)
```

Interpretation :  
25% of two door cars have more than a 130 Horse Power and 75% of two door cars have less than a 130 Horse Power

25% of four door cars have more than a 110 Horse Power and 75% of four door cars have less than a 110 Horse Power

the top line of the boxes represents the *Thirdquartile(Q3)*, and the base line is the *Firstdquartile(Q1)*, the Ts are called whiskers, it gives the smallest value present in the dataset or
(q1 - 1/2 * iqr), iqr = interquertile range can be declared as an outlier 
(q3 + 1/2 * iqr) will be declared as an outlier
the top whisker of two door cars can be considered as an outlier and the bottom one has the minimum value
(q3 - q1 = iqr)
The points outside of the whiskers are considered as outliers and those points are plotted individually

50% of the data lies inside the box
25% of the data lies above the box
25% of the data lies below the box

Since the data represented is pretty far higher in both the cases it can be said that the dist. is skewed to the right

```python
sns.boxplot(auto["number_of_doors"], auto["horsepower"], hue=auto['fuel_type']);
```

blue box plots - Gasoline cars, orange box plots - diesel cars

*BAR PLOTS: Plots the avg of a dataset, the black lines represents error bars,computes a 95% confidence interval for where the avg would be* 

```python
sns.barplot(auto['body_style'], auto['horsepower'],hue=auto['fuel_type']);
```

Instead of plotting only the avg of horsepowers, we can count the number of cars in each body type

```python
sns.countplot(auto['body_style'], hue=auto['fuel_type']);
```

*Multi-panel Categorical Plot : CAT PLOT: Makes multiple charts where each chart represents categorical values*

```python
sns.catplot(x='fuel_type',
            y='horsepower',
            hue='number_of_doors',
            col='drive_wheels',
            data=auto,
            kind='box');
```

```python
sns.catplot(x='fuel_type',
            y='horsepower',
            hue='number_of_doors',
            col='drive_wheels',
            data=auto,
            kind='point');
```

```python
sns.catplot(x='fuel_type',
            y='horsepower',
            hue='number_of_doors',
            col='drive_wheels',
            data=auto,
            kind='violin');  #KDE of horsepower of the cars
```

```python
sns.catplot(x='fuel_type',
            y='horsepower',
            hue='number_of_doors',
            col='drive_wheels',
            data=auto,
            kind='bar');
```

```python
sns.catplot(x='fuel_type',
            y='horsepower',
            hue='number_of_doors',
            col='drive_wheels',
            data=auto,
            kind='strip');
```

```python
sns.catplot(x='fuel_type',
            #y='horsepower',
            hue='number_of_doors',
            col='drive_wheels',
            data=auto,
            kind='count'); #y in a count plot shows the no. 
                           #of variables, so remove y  
```

```python
sns.lmplot(y="horsepower", x="engine_size", hue="fuel_type", data=auto);
```
